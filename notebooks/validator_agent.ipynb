```python
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool, ScrapeWebsiteTool
from langchain_google_genai import ChatGoogleGenerativeAI
import yaml
import os
```

```python
from crewai import LLM

llm = LLM(
    model = "gemini/gemini-2.5-flash",
    temperature=0.1, 
    api_key="${GOOGLE_API_KEY}" 
)

llm.call("Who is Ariana Grande")
```

```python
validator = Agent(
    role="Domain Knowledge Expert",
    goal="Check whether the given prompt is related to medicine and medicine related market",
    backstory="""You are an expert drug researcher with years of experience""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
def create_validator_crew(topic:str):
    validation_task=Task(
        description=f"""Your task is to determine whether the given {topic} is relevant to the pharmaceutical and drug development domain. Consider whether it relates to any of the following areas:

Drug discovery and development (molecular targets, drug candidates, preclinical studies)

Clinical trials (Phase I-IV studies, trial designs, patient recruitment)

Pharmaceutical market research (market size, therapy areas, competitive landscape, sales trends)

Intellectual property (patents, exclusivity, patent filings, licensing)

Regulatory aspects (FDA, EMA, or other health authority approvals and guidelines)

Return a clear yes/no if the topic is relevant to these areas and optionally provide a brief justification.""",
        agent=validator,
        expected_output="""A simple yes/no answer"""
    )

     # Create the crew
    crew = Crew(
        agents=[validator],
        tasks=[validation_task],
        process=Process.sequential,
        verbose=True
    )
    
    return crew
```

```python
validator_crew=create_validator_crew("Tbh I am not a good boy")
validator_crew.kickoff()
```
