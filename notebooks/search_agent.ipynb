```python
web_researcher = Agent(
    role="Web Intelligence Researcher",
    goal="Search the web for accurate and relevant information on given topics",
    backstory="""You are an expert web researcher with years of experience in 
    gathering intelligence from online sources. You know how to find credible 
    information, verify sources, and extract key insights from web content.""",
    tools=[SerperDevTool()],
    llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
data_analyst = Agent(
    role="Data Analyst",
    goal="Analyze gathered web intelligence and extract meaningful insights",
    backstory="""You are a skilled data analyst who specializes in processing 
    web intelligence data. You can identify patterns, trends, and key information 
    from various sources and synthesize them into actionable insights.""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
report_writer = Agent(
    role="Intelligence Report Writer",
    goal="Create comprehensive, well-structured intelligence reports",
    backstory="""You are a professional intelligence report writer with expertise 
    in creating clear, concise, and actionable reports. You can take complex 
    information and present it in an easy-to-understand format.""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
def create_web_intel_crew(topic: str):
    """
    Create a crew to gather web intelligence on a given topic
    
    Args:
        topic: The topic to research
    
    Returns:
        Crew: Configured CrewAI crew
    """
    
    # Task 1: Web Research
    research_task = Task(
        description=f"""
        Conduct comprehensive web research on the following topic: {topic}
        
        Your research should include:
        1. Search for the latest information and news
        2. Identify credible sources and authoritative websites
        3. Scrape relevant web pages for detailed information
        4. Gather facts, statistics, and expert opinions
        5. Note the sources and URLs for all information
        
        Provide a detailed summary of your findings.
        """,
        agent=web_researcher,
        expected_output="A comprehensive research summary with sources and URLs"
    )
    
    # Task 2: Data Analysis
    analysis_task = Task(
        description=f"""
        Analyze the research findings on {topic} and extract key insights.
        
        Your analysis should:
        1. Identify the most important findings
        2. Highlight trends and patterns
        3. Compare different perspectives or sources
        4. Identify any gaps in information
        5. Provide context and implications
        
        Create a structured analysis of the gathered intelligence.
        """,
        agent=data_analyst,
        expected_output="A structured analysis with key insights and implications",
        context=[research_task]
    )
    
    # Task 3: Report Writing
    report_task = Task(
        description=f"""
        Create a comprehensive intelligence report on {topic}.
        
        The report should include:
        1. Executive Summary
        2. Key Findings
        3. Detailed Analysis
        4. Sources and References
        5. Recommendations (if applicable)
        
        The report should be professional, well-structured, and actionable.
        """,
        agent=report_writer,
        expected_output="A complete intelligence report in a professional format",
        context=[research_task, analysis_task]
    )
    
    # Create the crew
    crew = Crew(
        agents=[web_researcher, data_analyst, report_writer],
        tasks=[research_task, analysis_task, report_task],
        process=Process.sequential,
        verbose=True
    )
    
    return crew
```

```python
def run_web_intel_agent(topic: str):
    """
    Run the web intelligence agent on a given topic
    
    Args:
        topic: The topic to research
    
    Returns:
        str: The final intelligence report
    """
    print(f"\n{'='*60}")
    print(f"Starting Web Intelligence Gathering on: {topic}")
    print(f"{'='*60}\n")
    
    crew = create_web_intel_crew(topic)
    result = crew.kickoff()
    
    print(f"\n{'='*60}")
    print("Intelligence Report Complete!")
    print(f"{'='*60}\n")
    
    return result
```

```python
def main(input):
    topic = f"{input}"
    
    # Run the web intelligence agent
    report = run_web_intel_agent(topic)
    
    # Print the final report
    print("\n" + "="*60)
    print("FINAL INTELLIGENCE REPORT")
    print("="*60 + "\n")
    print(report)
    
    # Optional: Save report to file
    with open("intelligence_report.txt", "w") as f:
        f.write(str(report))
    print("\nReport saved to intelligence_report.txt")
```
