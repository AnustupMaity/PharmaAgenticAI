```python
dotenv.load_dotenv()
```

```python
# === 1. LLM Configuration (Gemini) ===
llm = LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.2,
    api_key="${GOOGLE_API_KEY}",
    timeout=60
)
llm.call("Hello")
```

```python
# Read from .env
mongo_uri = os.getenv("MONGO_URI")
db_name = os.getenv("MONGO_DB")
collection_name = os.getenv("MONGO_COLLECTION")
```

```python
print(f"Mongo URI: {mongo_uri}")
print(f"Database Name: {db_name}")
print(f"Collection Name: {collection_name}")
```

```python
# === 2. Connect to MongoDB ===
client = MongoClient(mongo_uri)
db = client[db_name]
iqvia_collection = db[collection_name]

data = list(iqvia_collection.find({}, {"_id": 0}))
df = pd.DataFrame(data)
print("Loaded IQVIA data:")
display(df.head())
```

```python
# === 3. Agent Definitions ===

market_analyst = Agent(
    role="Market Data Analyst",
    goal="Extract and summarize quantitative insights from IQVIA market data.",
    backstory="""You are a pharmaceutical data analyst skilled in analyzing
    therapy-level sales and market share data to uncover market dynamics.""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)

trend_specialist = Agent(
    role="Trend Specialist",
    goal="Identify growth trends, CAGRs and therapy-level competition from IQVIA data.",
    backstory="""You are an expert in interpreting multi-year pharmaceutical data 
    to compute growth, volume shifts and competition intensity.""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)

insights_reporter = Agent(
    role="Insights Reporter",
    goal="Prepare an executive summary report of IQVIA analysis findings.",
    backstory="""You prepare crisp, data-backed insight summaries for business decisions.""",
    llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
# === 4. Task Definitions ===

task_extract = Task(
    description=(
        "Analyze the IQVIA dataset to determine market size by therapy, "
        "top countries by sales, and overall sales share distribution."
    ),
    expected_output=(
        "A concise table of therapies with total sales, market share %, "
        "and country-wise distribution."
    ),
    agent=market_analyst
)

task_trends = Task(
    description=(
        "Based on extracted data, calculate CAGR and identify sales growth/decline "
        "patterns per therapy and region. Highlight top 3 growing therapies."
    ),
    expected_output="CAGR values per therapy and bullet-point trend summary.",
    agent=trend_specialist,
    context=[task_extract]
)

task_report = Task(
    description=(
        "Generate a final business insights report that includes:\n"
        "1. Market size summary\n2. CAGR trends\n3. Competitive landscape\n"
        "4. Key recommendations."
    ),
    expected_output="Well-structured business report with tabular and narrative sections.",
    agent=insights_reporter,
    context=[task_extract, task_trends]
)
```

```python
from datetime import datetime
from bson import ObjectId

def sanitize_inputs(data):
    """Convert unsupported types (datetime, ObjectId) to strings."""
    if isinstance(data, list):
        return [sanitize_inputs(d) for d in data]
    elif isinstance(data, dict):
        return {k: sanitize_inputs(v) for k, v in data.items()}
    elif isinstance(data, datetime):
        return data.isoformat()
    elif isinstance(data, ObjectId):
        return str(data)
    else:
        return data
```

```python
# === 5. Crew Definition ===
crew = Crew(
    agents=[market_analyst, trend_specialist, insights_reporter],
    tasks=[task_extract, task_trends, task_report],
    process=Process.sequential,
    verbose=False
)
```

```python
def run_iqvia_agent(topic):
    print(f"\n{'='*60}")
    print(f"Starting IQVIA Insights Analysis on: {topic}")
    print(f"{'='*60}\n")

    # Fetch data from MongoDB
    iqvia_data = list(iqvia_collection.find())
    print(f"Loaded {len(iqvia_data)} records from MongoDB")

    # Sanitize datetime/ObjectId
    clean_data = sanitize_inputs(iqvia_data)
    print(clean_data)

    # Run the crew
    result = crew.kickoff(inputs={"topic": topic, "iqvia_data": clean_data})

    print(f"\n{'='*60}")
    print("IQVIA Insights Report Complete!")
    print(f"{'='*60}\n")

    print(result)
    with open("iqvia_insights_report.txt", "w") as f:
        f.write(str(result))
    print("Report saved to iqvia_insights_report.txt")
```

```python
run_iqvia_agent("Compare oncology vs cardiology sales performance")
```
