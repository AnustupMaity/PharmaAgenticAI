```python
from crewai import Agent, Crew, Task, Process
from crewai_tools import PDFSearchTool
from crewai import LLM
from IPython.display import display, clear_output
import ipywidgets as widgets
import os
```

```python
# Create widgets for interactive interface
upload = widgets.FileUpload(accept='.pdf', multiple=False)
question_input = widgets.Text(
    placeholder="Ask a question about the PDF...",
    description="Question:",
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)
ask_button = widgets.Button(
    description="Ask Question",
    button_style='primary'
)
output_area = widgets.Output()

# Display upload widget
print("ðŸ“š PDF Q&A Tool")
print("Upload a PDF and ask questions about it!")
display(upload)
```

```python
llm = LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.1, 
    api_key="${GOOGLE_API_KEY}"
)

# Test LLM
print("ðŸ¤– Testing LLM connection...")
test_response = llm.call("Say hello!")
print(f"âœ… LLM working: {test_response}")
```

```python
# Global variables
current_agent = None
current_pdf_path = None

def process_pdf(change):
    global current_agent, current_pdf_path
    
    with output_area:
        clear_output()
        print("ðŸ“„ Processing uploaded PDF...")
        
        try:
            if not upload.value:
                print("âŒ No file uploaded")
                return
                
            # Get the uploaded file
            for name, file_info in upload.value.items():
                current_pdf_path = name
                
                # Save the uploaded PDF locally
                with open(name, 'wb') as f:
                    f.write(file_info['content'])
                print(f"âœ… Saved PDF: {name}")

                # Initialize the PDF search tool
                print("ðŸ” Setting up PDF search tool...")
                pdf_tool = PDFSearchTool(pdf=name)

                # Create a CrewAI agent for answering questions
                print("ðŸ§  Creating AI agent...")
                current_agent = Agent(
                    role="PDF Document Analyst",
                    goal="Answer questions accurately based on the content of the uploaded PDF document",
                    backstory="You are an expert document analyst who can thoroughly read and understand PDF documents. You provide detailed, accurate answers based on the document content.",
                    llm=llm,
                    tools=[pdf_tool],
                    verbose=True
                )

                print("ðŸŽ‰ Setup complete! You can now ask questions about the PDF.")
                
                # Show the question interface
                display(widgets.VBox([
                    widgets.HTML("<h3>Ask a question about your PDF:</h3>"),
                    question_input,
                    ask_button
                ]))
                
                break
                
        except Exception as e:
            print(f"âŒ Error processing PDF: {e}")
            print(f"Error type: {type(e).__name__}")

def ask_question(button):
    global current_agent
    
    with output_area:
        if current_agent is None:
            print("âŒ Please upload a PDF first!")
            return
            
        question = question_input.value.strip()
        if not question:
            print("âŒ Please enter a question!")
            return
            
        print(f"ðŸ¤” Question: {question}")
        print("ðŸ” Searching PDF and generating answer...")
        
        try:
            # Create a task for the agent
            task = Task(
                description=f"Answer this question based on the PDF content: {question}",
                expected_output="A comprehensive answer based on the PDF document",
                agent=current_agent
            )
            
            # Create crew and execute
            crew = Crew(
                agents=[current_agent],
                tasks=[task],
                process=Process.sequential,
                verbose=False
            )
            
            result = crew.kickoff()
            
            print("\n" + "="*50)
            print("ðŸ“ Answer:")
            print("="*50)
            print(result.raw if hasattr(result, 'raw') else str(result))
            
            # Clear the input for next question
            question_input.value = ""
            
        except Exception as e:
            print(f"âŒ Error generating answer: {e}")
            print("Please try rephrasing your question.")

# Connect the button to the function
ask_button.on_click(ask_question)
```

```python
# Connect the upload widget to the processing function
upload.observe(process_pdf, names='value')

# Display the output area where results will show
display(output_area)
```

```python
# Simplified PDF Q&A - Direct approach
import tempfile

def simple_pdf_qa():
    """Simple direct approach for PDF Q&A"""
    
    if not upload.value:
        print("âŒ Please upload a PDF file first!")
        return
    
    print("ðŸ“„ Processing PDF...")
    
    try:
        # Get the uploaded file
        filename, file_info = next(iter(upload.value.items()))
        
        # Create a temporary file
        with tempfile.NamedTemporaryFile(suffix='.pdf', delete=False) as temp_file:
            temp_file.write(file_info['content'])
            temp_path = temp_file.name
        
        print(f"âœ… Saved temporary PDF: {temp_path}")
        
        # Create PDF search tool
        pdf_tool = PDFSearchTool(pdf=temp_path)
        print("ðŸ” PDF tool created")
        
        # Create agent
        agent = Agent(
            role="PDF Assistant",
            goal="Answer questions about the PDF",
            backstory="You are a helpful assistant that answers questions based on PDF content.",
            llm=llm,
            tools=[pdf_tool],
            verbose=True
        )
        print("ðŸ¤– Agent created")
        
        # Test question
        test_question = "What is this document about?"
        print(f"ðŸ¤” Testing with question: {test_question}")
        
        # Create and run task
        task = Task(
            description=f"Answer: {test_question}",
            expected_output="A summary of the document content",
            agent=agent
        )
        
        crew = Crew(agents=[agent], tasks=[task], verbose=True)
        result = crew.kickoff()
        
        print("\n" + "="*40)
        print("ðŸ“ RESULT:")
        print("="*40)
        print(result.raw if hasattr(result, 'raw') else str(result))
        
        # Cleanup
        os.unlink(temp_path)
        
    except Exception as e:
        print(f"âŒ Error: {e}")
        import traceback
        traceback.print_exc()

# Run the simple test
simple_pdf_qa()
```

```python
# Direct Gemini PDF Processing - Bypass PDFSearchTool
import PyPDF2
import tempfile
from typing import List

def extract_text_from_pdf(pdf_path: str) -> str:
    """Extract text from PDF using PyPDF2"""
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
    except Exception as e:
        print(f"Error extracting PDF text: {e}")
    return text

def chunk_text(text: str, chunk_size: int = 2000, overlap: int = 200) -> List[str]:
    """Split text into overlapping chunks"""
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start = end - overlap
    return chunks

def gemini_pdf_qa():
    """Direct Gemini PDF Q&A without PDFSearchTool"""
    
    if not upload.value:
        print("âŒ Please upload a PDF file first!")
        return
    
    print("ðŸ“„ Starting Direct Gemini PDF Processing...")
    
    try:        
        # Handle different upload formats
        if isinstance(upload.value, tuple):
            file_info = upload.value[0]
            filename = file_info.get('name', 'uploaded_file.pdf')
            content = file_info.get('content', b'')
        elif isinstance(upload.value, dict):
            filename, file_info = next(iter(upload.value.items()))
            content = file_info['content']
        else:
            raise ValueError(f"Unsupported upload format: {type(upload.value)}")
            
        print(f"Processing: {filename} ({len(content)} bytes)")
        
        # Save PDF temporarily
        with tempfile.NamedTemporaryFile(suffix='.pdf', delete=False) as temp_file:
            temp_file.write(content)
            temp_path = temp_file.name
        
        print("âœ… PDF saved temporarily")
        
        # Extract text from PDF
        print("ðŸ“– Extracting text from PDF...")
        pdf_text = extract_text_from_pdf(temp_path)
        
        if not pdf_text.strip():
            print("âŒ No text could be extracted from the PDF")
            return
            
        print(f"âœ… Extracted {len(pdf_text)} characters of text")
        print(f"Preview: {pdf_text[:200]}...")
        
        # Chunk the text
        chunks = chunk_text(pdf_text)
        print(f"âœ… Split into {len(chunks)} text chunks")
        
        # Create context from first few chunks (for demonstration)
        context = "\n".join(chunks[:3])  # Use first 3 chunks
        
        # Test question
        test_question = "What is this document about? Provide a summary."
        
        # Create prompt for Gemini
        prompt = f"""Based on the following document content, please answer the question.

Document Content:
{context}

Question: {test_question}

Please provide a comprehensive answer based only on the information in the document."""

        print(f"ðŸ¤” Asking Gemini: {test_question}")
        
        # Query Gemini directly
        response = llm.call(prompt)
        
        print("\n" + "="*50)
        print("ðŸŽ‰ GEMINI DIRECT RESPONSE:")
        print("="*50)
        print(response)
        
        # Cleanup
        os.unlink(temp_path)
        print("\nâœ… Cleanup completed")
        
        return response
        
    except Exception as e:
        print(f"\nâŒ ERROR in direct Gemini processing: {e}")
        import traceback
        traceback.print_exc()
        
        # Cleanup
        try:
            if 'temp_path' in locals():
                os.unlink(temp_path)
        except:
            pass

# Run the direct Gemini approach
gemini_pdf_qa()
```
